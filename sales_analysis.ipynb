```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.isnull().sum()
```

```python
df['review_rating']=df.groupby("category")["review_rating"].transform(lambda x:x.fillna(x.median()))
```

```python
r=['young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=r)
```

```python
labels={"Fortnightly":14,"Weekly":7,"Monthly":30,"Quarterly":90,"Bi-Weekly":14,"Annually":365,"Every 3 Months":90}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(labels)
```

```python
from sqlalchemy import create_engine

server = 'localhost\SQLEXPRESS'      # e.g., DESKTOP-XXXX\SQLEXPRESS
database = 'Sales'

engine = create_engine(
    f"mssql+pyodbc://{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server"
)
```
